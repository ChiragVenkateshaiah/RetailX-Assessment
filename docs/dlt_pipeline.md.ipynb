## Delta Live Table (DLT)
To make the pipeline production-ready, Gold tables were implemented using Delta Live Tables (DLT).

### Why DLT?
- Declarative pipeline definitions
- Built-in data quality enforcement
- Automatic dependency management
- Lineage and observability out of the box

### DLT Implementation
- Pipeline type: ETL Pipeline
- Model: Triggered
- Product edition: Core
- Target schema: `dlt_gold`

### Data Quality Expectations
Examples:
- Order date must not be null
- Revenue values must be positive
- Customer country must be present

These expectations are defined using `@dlt.expect` decorators.

> Delta Live Tables enable declarative pipelines with built-in data quality checks, monitoring, and automatic orchestration, making pipelines production-ready
